```python
from transformers import AutoProcessor, SeamlessM4Tv2Model
import torchaudio

# Initialize processor and model
processor = AutoProcessor.from_pretrained("facebook/seamless-m4t-v2-large")
model = SeamlessM4Tv2Model.from_pretrained("facebook/seamless-m4t-v2-large")

# Load and process audio
audio, orig_freq = torchaudio.load("/content/temp3.mp3")
audio = torchaudio.functional.resample(audio, orig_freq=orig_freq, new_freq=16000)

# Get audio inputs
inputs = processor(audios=audio, return_tensors="pt", sampling_rate=16000)

# Generate text outputs
# For transcription (source language to text)
output_tokens = model.generate(**inputs, tgt_lang="hin", generate_speech=False)
transcribed_text = processor.decode(output_tokens[0].cpu().numpy().squeeze(), skip_special_tokens=True)

print("Transcribed Text (Original Language):")
print(transcribed_text)
```

```python
# For translation (source language to target language text)
output_tokens_eng = model.generate(**inputs, tgt_lang="eng", generate_speech=False)
translated_text = processor.decode(output_tokens_eng[0].cpu().numpy().squeeze(), skip_special_tokens=True)

print("\nTranslated Text (English):")
print(translated_text)
```
