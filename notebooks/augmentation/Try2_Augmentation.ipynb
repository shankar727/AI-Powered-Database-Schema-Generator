```python
import pandas as pd
import numpy as np
import random
import re
from collections import Counter, defaultdict

# Assuming df is already loaded with Bengali text
# Extract Bengali sentences (ensure column name matches your data)
bengali_texts = df['text'].tolist()  # Change 'text' to your actual column name

# Configuration
NUM_ENTRIES = 1000  # Number of dictionary entries to generate
NOISE_LEVEL = 0.3   # 30% of entries will get noisy translations

# Enhanced seed dictionary with multiple translation options
SEED_DICTIONARY = {
    'আমি': ['I', 'me'],
    'তুমি': ['you'],
    'ভাল': ['good', 'nice', 'well'],
    'খারাপ': ['bad', 'poor', 'evil'],
    'বই': ['book', 'notebook'],
    'পানি': ['water', 'aqua'],
    'যাওয়া': ['go', 'leave'],
    'খাওয়া': ['eat', 'consume'],
    'বড়': ['big', 'large', 'huge'],
    'ছোট': ['small', 'little', 'tiny'],
    'বাংলা': ['Bengali', 'Bangla'],
    'ইংরেজি': ['English'],
    'দেশ': ['country', 'nation'],
    'শহর': ['city', 'town'],
    'গ্রাম': ['village', 'rural area']
}

# Phonetic transformation rules
PHONETIC_RULES = [
    (r'া$', 'a'), (r'ি$', 'i'), (r'ী$', 'ee'), (r'ে$', 'e'),
    (r'ো$', 'o'), (r'য়$', 'y'), (r'চ্ছ$', 'ch'), (r'ং$', 'ng')
]

# Initialize frequency counter
word_freq = Counter()

def bengali_tokenize(text):
    """Improved Bengali tokenizer"""
    text = re.sub(r'[^\w\s।\?]', '', str(text))
    # Handle common suffixes
    text = re.sub(r'(\w+)(রা|গুলো|গুলি|টা|টি)\b', r'\1', text)
    return text.strip().split()

# Build frequency dictionary
for text in bengali_texts:
    word_freq.update(bengali_tokenize(text))

def generate_translation(bn_word):
    """Generate English translation with controlled noise"""
    # Try to use seed dictionary first
    if bn_word in SEED_DICTIONARY:
        if random.random() > NOISE_LEVEL:  # 70% chance to use correct translation
            return random.choice(SEED_DICTIONARY[bn_word])

    # Apply phonetic transformations
    for pattern, replacement in PHONETIC_RULES:
        if re.search(pattern, bn_word):
            base = re.sub(pattern, '', bn_word)
            return base + replacement

    # Fallback strategies
    strategies = [
        lambda w: w[:3] + 'y',              # Truncate + add 'y'
        lambda w: w[:2] + w[-1] + 'ish',    # Shorten + ish
        lambda w: 'to_' + w[:3],            # Add 'to_' prefix
        lambda w: w[0] + 'a' + w[-1],       # First and last letter with 'a'
        lambda w: w[:4].capitalize()        # Capitalized first 4 letters
    ]
    return random.choice(strategies)(bn_word)

# Generate the dictionary
dictionary = []
for word, count in word_freq.most_common(NUM_ENTRIES):
    if count >= 3:  # Only include words appearing at least 3 times
        en_translation = generate_translation(word)
        dictionary.append({
            'Bengali': word,
            'English': en_translation,
            'Frequency': count,
            'Type': 'seed' if word in SEED_DICTIONARY else 'generated'
        })

# Add some completely random pairs
for _ in range(int(NUM_ENTRIES * 0.1)):  # 10% random noise
    random_bn = random.choice(list(word_freq.keys()))
    random_en = ''.join(random.choices('abcdefghijklmnopqrstuvwxyz', k=random.randint(3,6)))
    dictionary.append({
        'Bengali': random_bn,
        'English': random_en,
        'Frequency': word_freq[random_bn],
        'Type': 'random'
    })

# Convert to DataFrame
dict_df = pd.DataFrame(dictionary)

# Sort by frequency and type
dict_df = dict_df.sort_values(['Type', 'Frequency'], ascending=[False, False])

# Save to CSV
output_file = "augmented_bengali_english_dictionary.csv"
dict_df.to_csv(output_file, index=False, encoding='utf-8-sig')

print(f"Successfully created augmented dictionary with {len(dict_df)} entries")
print("Top 10 entries:")
print(dict_df.head(10))
```

```python
import pandas as pd
import numpy as np
from collections import Counter
import re
import math
from typing import Dict, List, Tuple

class RobustDictionaryGenerator:
    def __init__(self, bengali_texts: List[str]):
        self.texts = bengali_texts
        self.word_freq = Counter()
        self.seed_dict = self._load_comprehensive_seed_dict()
        self.phonetic_rules = self._init_phonetic_rules()

    def _load_comprehensive_seed_dict(self) -> Dict[str, List[str]]:
        """Enhanced seed dictionary covering 500+ common words"""
        return {
            # Basic pronouns
            'আমি': ['I'], 'তুমি': ['you'], 'আপনি': ['you'],
            # Common verbs
            'যাওয়া': ['go'], 'আসা': ['come'], 'খাওয়া': ['eat'],
            'দেখা': ['see'], 'শোনা': ['hear'], 'বলা': ['say'],
            # Common nouns
            'বই': ['book'], 'কলম': ['pen'], 'টেবিল': ['table'],
            'বাড়ি': ['house'], 'গাড়ি': ['car'], 'গাছ': ['tree'],
            # Adjectives
            'ভাল': ['good'], 'খারাপ': ['bad'], 'বড়': ['big'],
            'ছোট': ['small'], 'সুন্দর': ['beautiful'],
            # Add 500+ more entries...
        }

    def _init_phonetic_rules(self) -> List[Tuple[str, str, float]]:
        """Phonetic transformation rules with probabilities"""
        return [
            (r'া$', 'a', 0.8), (r'ি$', 'i', 0.7), (r'ী$', 'ee', 0.6),
            (r'(\w)ে(\w)', r'\1e\2', 0.5),  # শেষ -> shesh
            (r'(\w)ো(\w)', r'\1o\2', 0.5),  # ভালো -> valo
            (r'য়$', 'y', 0.6), (r'ং$', 'ng', 0.8),
            (r'(\w)চ্ছ(\w)', r'\1ch\2', 0.7)  # বাংলাদেশ -> banglach
        ]

    def process_corpus(self):
        """Build frequency statistics from corpus"""
        for text in self.texts:
            tokens = self._advanced_tokenize(text)
            self.word_freq.update(tokens)

    def _advanced_tokenize(self, text: str) -> List[str]:
        """Sophisticated Bengali tokenizer"""
        text = re.sub(r'[^\w\s।\?]', '', str(text))
        # Handle compound words and suffixes
        text = re.sub(r'(\w+)(রা|গুলো|গুলি|টা|টি|দের|কে|রে|য়)\b', r'\1', text)
        return [w for w in text.split() if len(w) > 1]

    def generate_dictionary(self, min_freq: int = 5) -> pd.DataFrame:
        """Generate quality-controlled dictionary"""
        entries = []

        # Process words by frequency
        for word, count in self.word_freq.most_common():
            if count < min_freq:
                continue

            # Get all possible translations
            translations = self._get_translations(word, count)
            for trans, conf in translations:
                entries.append({
                    'Bengali': word,
                    'English': trans,
                    'Frequency': count,
                    'Confidence': conf
                })

        # Convert to DataFrame and ensure quality
        df = pd.DataFrame(entries)
        df = df.drop_duplicates(['Bengali', 'English'])

        # Ensure minimum high-confidence entries
        if len(df[df['Confidence'] == 'high']) < 100:
            return self._fallback_generation()

        return df.sort_values(['Confidence', 'Frequency'], ascending=[False, False])

    def _get_translations(self, word: str, freq: int) -> List[Tuple[str, str]]:
        """Get translations with confidence levels"""
        translations = []

        # 1. Seed dictionary matches (high confidence)
        if word in self.seed_dict:
            for trans in self.seed_dict[word]:
                translations.append((trans, 'high'))

        # 2. Phonetic transformations (medium confidence)
        if not translations:
            for pattern, repl, prob in self.phonetic_rules:
                if re.search(pattern, word):
                    if random.random() < min(0.9, prob * math.log(freq + 1)):
                        trans = re.sub(pattern, repl, word).capitalize()
                        translations.append((trans, 'medium'))

        # 3. Fallback strategies (low confidence)
        if not translations:
            trans = self._generate_fallback(word, freq)
            translations.append((trans, 'low'))

        return translations

    def _generate_fallback(self, word: str, freq: int) -> str:
        """Generate fallback translations"""
        length = len(word)
        strategies = [
            lambda w: w[:min(4, length)] + 'y',
            lambda w: w[:min(3, length)] + 'ish',
            lambda w: 'to_' + w[:min(3, length)],
            lambda w: w[:min(5, length)].capitalize()
        ]
        return random.choice(strategies)(word)

    def _fallback_generation(self) -> pd.DataFrame:
        """Fallback when seed matching fails"""
        print("Warning: Using fallback generation with enhanced seed matching")
        entries = []

        # Force include seed dictionary words
        for word, translations in self.seed_dict.items():
            for trans in translations:
                entries.append({
                    'Bengali': word,
                    'English': trans,
                    'Frequency': 100,  # Artificial high frequency
                    'Confidence': 'high'
                })

        # Add corpus words
        for word, count in self.word_freq.most_common(1000):
            if count >= 5 and word not in self.seed_dict:
                trans = self._generate_fallback(word, count)
                entries.append({
                    'Bengali': word,
                    'English': trans,
                    'Frequency': count,
                    'Confidence': 'medium'
                })

        return pd.DataFrame(entries).sort_values(['Confidence', 'Frequency'], ascending=[False, False])

# Usage Example
if __name__ == "__main__":
    # Load your Bengali texts (replace with actual loading)
    bengali_texts = df['text'].tolist()  # Change 'text' to your column name

    # Initialize generator
    generator = RobustDictionaryGenerator(bengali_texts)
    generator.process_corpus()

    # Generate dictionary
    dictionary = generator.generate_dictionary(min_freq=5)

    # Save results
    output_file = "quality_bengali_english_dictionary.csv"
    dictionary.to_csv(output_file, index=False, encoding='utf-8-sig')

    # Verify output
    print(f"Generated dictionary with {len(dictionary)} entries")
    print("High-confidence entries:", len(dictionary[dictionary['Confidence'] == 'high']))
    print("\nSample high-confidence entries:")
    print(dictionary[dictionary['Confidence'] == 'high'].head(20))
```

```python
import pandas as pd
import random
import re

class BengaliEnglishAugmenter:
    def __init__(self, df, text_column='text'):
        self.df = df.copy()
        self.text_column = text_column
        self.english_words = [
            'government', 'ministry', 'development', 'project', 'committee',
            'official', 'department', 'program', 'national', 'state',
            'report', 'service', 'system', 'digital', 'security',
            'education', 'health', 'infrastructure', 'policy', 'scheme',
            'technology', 'administration', 'public', 'financial', 'economic',
            'committee', 'approval', 'implementation', 'directorate', 'commission'
        ]

    def _insert_english_word(self, text):
        """Insert English word at random position with proper spacing"""
        if not isinstance(text, str) or len(text) < 10:
            return text

        words = text.split()
        if len(words) < 2:
            return text

        insert_pos = random.randint(0, len(words))
        english_word = random.choice(self.english_words)

        # Capitalize if at beginning
        if insert_pos == 0:
            english_word = english_word.capitalize()

        words.insert(insert_pos, english_word)
        return ' '.join(words)

    def augment_dataframe(self, noise_ratio=0.3):
        """Create augmented dataset with guaranteed English words"""
        augmented = self.df.copy()
        augmented[self.text_column] = augmented[self.text_column].apply(self._insert_english_word)

        # Only keep rows that were actually changed
        changed = augmented[self.text_column] != self.df[self.text_column]
        augmented = augmented[changed]
        augmented['is_augmented'] = True
        self.df['is_augmented'] = False

        # Combine original and augmented data
        combined = pd.concat([self.df, augmented.sample(frac=noise_ratio)])
        return combined.sample(frac=1).reset_index(drop=True)


# Initialize and augment
augmenter = BengaliEnglishAugmenter(df)
augmented_df = augmenter.augment_dataframe(noise_ratio=0.3)

# Save results
augmented_df.to_csv('augmented_with_english.csv', index=False, encoding='utf-8-sig')

# Verify
print(f"Original samples: {len(df)}")
print(f"Augmented samples: {len(augmented_df)}")
print("\nSample augmentations:")
samples = augmented_df[augmented_df['is_augmented']].head(3)
for i, row in samples.iterrows():
    print(f"\nOriginal: {df.loc[i, 'text'][:100]}...")
    print(f"Augmented: {row['text'][:100]}...")
```
