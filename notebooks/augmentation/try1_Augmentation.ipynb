```python
from datasets import load_dataset

dataset = load_dataset("ai4bharat/sangraha", "synthetic", split="ben")
```

```python
import pandas as pd
import re
import nltk
from collections import Counter

# Download NLTK tokenizer
nltk.download('punkt')

# Load parallel corpus
df = pd.read_csv("parallel_corpus.csv")

# Clean and tokenize
def tokenize(text):
    text = re.sub(r'[^\w\s]', '', text)
    tokens = text.strip().split()
    return tokens

def english_tokenize(text):
    tokens = nltk.word_tokenize(text)
    return [t.lower() for t in tokens]

# Build word alignment candidates
candidate_pairs = []

for index, row in df.iterrows():
    hindi_sentence = row['hindi']
    english_sentence = row['english']

    hindi_tokens = tokenize(hindi_sentence)
    english_tokens = english_tokenize(english_sentence)

    # Very naive alignment assumption (co-occurrence within sentence pair)
    for h_word in hindi_tokens:
        for e_word in english_tokens:
            candidate_pairs.append( (h_word, e_word) )

# Count co-occurrence frequencies
pair_counter = Counter(candidate_pairs)

# Filter strong candidates (can tune threshold)
min_count_threshold = 3  # Only keep pairs that co-occur at least 3 times
final_dict = {}

for (hindi_word, english_word), count in pair_counter.items():
    if count >= min_count_threshold:
        if hindi_word not in final_dict:
            final_dict[hindi_word] = english_word

# Save bilingual dictionary
dict_df = pd.DataFrame(list(final_dict.items()), columns=['Hindi', 'English'])
dict_df.to_csv("bilingual_dictionary.csv", index=False, encoding='utf-8-sig')

print("✅ Bilingual dictionary built successfully. File: bilingual_dictionary.csv")
```
