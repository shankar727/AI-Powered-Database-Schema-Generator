```python
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = HuggingFacePipeline.from_model_id(
    model_id='google/gemma-3-4b-it',
    task='text-generation',
    pipeline_kwargs=dict(
        temperature=0.5,
        max_new_tokens=3000
    )
)

model = ChatHuggingFace(llm=llm)

prompt = PromptTemplate(
    input_variables=["input_text"],
    template="""
You are a language expert skilled in Bengali-English code-mixing (Benglish). Your task is to rewrite the given Bengali sentence by naturally blending 20% to 40% English words into it, while keeping the sentence grammatically correct and semantically equivalent to the original. Follow these rules strictly:

1. The meaning and overall structure (syntax) of the original Bengali sentence must remain unchanged.
2. Replace or insert English words that are **commonly used or acceptable** in informal spoken or written BenEnglish (e.g., school, problem, plan, idea, smart, important, etc.).
3. Ensure the resulting sentence feels **natural, fluent, and idiomatic** to a bilingual Bengali speaker who frequently switches between Bengali and English in daily conversation.
4. Do **not** translate the entire sentence to English. The code-mixed version should still primarily be in Bengali (60%–80%) with the remaining 20%–40% in English.
5. Avoid awkward or forced insertions; prefer English words for abstract nouns, tech terms, academic terms, modern expressions, etc.

Example Input:
আমি আগামীকাল একটি গুরুত্বপূর্ণ মিটিংয়ে অংশগ্রহণ করব।

Example Output:
আমি কাল একটা important মিটিং attend করব।

Now, code-mix this Bengali sentence as per the instructions:
{input_text}
"""
)

parser = StrOutputParser()

chain = prompt | model | parser

result = chain.invoke({'input_text': """
বইমেলায় অনেক মানুষ এসেছে। নতুন নতুন বই দেখা যাচ্ছে। কিছু লেখকের সঙ্গে পরিচয়ও হলো। আমি কয়েকটি গল্পের বই কিনলাম। বন্ধুদের সঙ্গে দেখা করে আমরা একসঙ্গে খেতে গেলাম।
"""})

print(result)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch, pandas as pd
from tqdm import tqdm
import os
import gc
import psutil

# ⚙️ Config
MODEL_NAME = "google/mt5-small"  # Using smallest available model
OUTPUT_CSV = "dataset_code_mixed.csv"
CHECKPOINT = "checkpoint.txt"
MAX_TOKENS = 60
MAX_INPUT_TOKENS = 64  # Reduced from 128
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# 🧹 Memory monitoring
def print_memory_usage():
    process = psutil.Process(os.getpid())
    print(f"Memory usage: {process.memory_info().rss / 1024 / 1024:.2f} MB")
    if torch.cuda.is_available():
        print(f"GPU memory: {torch.cuda.memory_allocated() / 1024 / 1024:.2f} MB / {torch.cuda.max_memory_allocated() / 1024 / 1024:.2f} MB")

# 🔥 Load model with 8-bit quantization
try:
    tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
    model = AutoModelForSeq2SeqLM.from_pretrained(
        MODEL_NAME,
        device_map="auto",
        load_in_8bit=True,  # 8-bit quantization
        torch_dtype=torch.float16
    )
    print("Model loaded in 8-bit precision")
except Exception as e:
    print(f"8-bit loading failed: {e}")
    # Fallback to regular loading
    model = AutoModelForSeq2SeqLM.from_pretrained(
        MODEL_NAME,
        torch_dtype=torch.float16
    ).to(DEVICE)
    print("Model loaded in 16-bit precision")

model.eval()  # Disable dropout

# 📄 Load dataset
# Ensure df is loaded and filtered before this cell
if 'df' in locals():
    # Reset index after filtering to ensure a simple 0-based index
    df = df.reset_index(drop=True)
    df["code_mixed"] = ""  # Initialize column
    n = len(df)
else:
    print("DataFrame 'df' not found. Please load and filter the dataset first.")
    n = 0 # Set n to 0 to prevent loop from running

# ➡️ Resume from checkpoint
start_idx = 0
if os.path.exists(CHECKPOINT):
    with open(CHECKPOINT, 'r') as f:
        start_idx = int(f.read().strip())
    print(f"Resuming from row {start_idx}...")

    if os.path.exists(OUTPUT_CSV):
        try:
            existing_df = pd.read_csv(OUTPUT_CSV)
            if len(existing_df) >= start_idx:
                # Ensure existing_df also has a simple index or align by content if necessary
                # For simplicity, assuming the checkpoint and output CSV correspond to the reset index
                # A more robust approach would align based on unique identifiers if available
                 df.iloc[:start_idx] = existing_df.iloc[:start_idx]
        except pd.errors.EmptyDataError:
             print(f"Checkpoint file {CHECKPOINT} found, but {OUTPUT_CSV} is empty. Starting from the beginning.")
             start_idx = 0


# ⏰ Process with aggressive memory management
with torch.no_grad():
    for i in tqdm(range(start_idx, n)):
        try:
            # Clear memory aggressively
            torch.cuda.empty_cache()
            gc.collect()

            # Access row by integer position after resetting index
            text = df.loc[i, "tgt_txt"]
            prompt = f"Code-mix this Bengali sentence naturally with English, preserving meaning: {text}"

            # Tokenize with strict length limits
            inputs = tokenizer(
                prompt,
                return_tensors="pt",
                max_length=MAX_INPUT_TOKENS,
                truncation=True,
                padding="max_length"
            )

            # Move to device (with fallback to CPU if needed)
            try:
                inputs = {k: v.to(DEVICE) for k, v in inputs.items()}
            except RuntimeError as e:
                print(f"Moving to GPU failed, falling back to CPU: {e}")
                DEVICE = "cpu"
                model = model.to(DEVICE)
                inputs = {k: v.to(DEVICE) for k, v in inputs.items()}

            # Generate with memory optimizations
            try:
                output = model.generate(
                    **inputs,
                    max_new_tokens=MAX_TOKENS,
                    do_sample=True,
                    temperature=0.9,
                    num_beams=1,  # Disable beam search
                    use_cache=False,
                    early_stopping=True
                )

                decoded = tokenizer.decode(output[0], skip_special_tokens=True)
                print(f"Row {i}: {decoded}") # Print row number and output
                # Set value using integer position after resetting index
                df.loc[i, "code_mixed"] = decoded

            except RuntimeError as e:
                print(f"Generation failed for row {i}: {e}")
                df.loc[i, "code_mixed"] = f"[ERROR: GENERATION FAILED: {e}]"

            # Save progress after each row
            # Ensure header is written only once
            header = not os.path.exists(OUTPUT_CSV)
            df.iloc[[i]].to_csv(OUTPUT_CSV, mode='a', header=header, index=False)

            # Update checkpoint
            with open(CHECKPOINT, 'w') as f:
                f.write(str(i + 1))

            # Clear variables
            del inputs, output, decoded
            torch.cuda.empty_cache()
            gc.collect()

        except Exception as e:
            print(f"Error processing row {i}: {str(e)}")
            df.loc[i, "code_mixed"] = f"[ERROR: {str(e)}]"
            header = not os.path.exists(OUTPUT_CSV)
            df.iloc[[i]].to_csv(OUTPUT_CSV, mode='a', header=header, index=False)
            with open(CHECKPOINT, 'w') as f:
                f.write(str(i + 1))
            continue

# Clean up
if os.path.exists(CHECKPOINT):
    os.remove(CHECKPOINT)

print("✅ Code mixing completed!")
print_memory_usage()
```

```python
# Configuration
MODEL_NAME = "Qwen/Qwen2.5-Coder-3B-Instruct"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
BATCH_SIZE = 1  # Reduce if OOM errors occur
MAX_LENGTH = 256
REPLACE_RATIO = 0.35 # % of content words to replace
```

```python
# Load semantic similarity model (for validation)
semantic_model = SentenceTransformer('sentence-transformers/LaBSE', device=DEVICE)
```

```python
# Load Qwen model with 4-bit quantization for memory efficiency
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, use_fast=False)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    attn_implementation="flash_attention_2"
)
```

```python
def generate_code_mixed_sentence(bn_sentence: str) -> str:
    """Generate code-mixed version using Qwen with instruction prompting"""
    prompt = f"""তুমি একজন বাংলা-ইংলিশ কোড-মিক্সিং বিশেষজ্ঞ। নিচের বাংলা বাক্যটিকে বাংলা এবং ইংরেজির মিশ্রণে রূপান্তর কর,
    যেখানে ৩০-৪০% কন্টেন্ট শব্দ (বিশেষ্য, ক্রিয়া, বিশেষণ) ইংরেজিতে রূপান্তর করবে। কার্যকরী শব্দ (অব্যয়, সর্বনাম, বিভক্তি) বাংলায় রাখবে।
    অর্থ অপরিবর্তিত রাখতে হবে। শুধুমাত্র রূপান্তরিত বাক্যটি আউটপুট দেবে।

    ইনপুট: {bn_sentence}
    আউটপুট:"""

    response = generator(
        prompt,
        num_return_sequences=1,
        pad_token_id=tokenizer.eos_token_id
    )

    # Extract generated text after prompt
    return response[0]['generated_text'].split("আউটপুট:")[-1].strip()

def validate_semantics(original: str, mixed: str, threshold=0.75) -> bool:
    """Ensure semantic similarity using LaBSE embeddings"""
    emb_orig = semantic_model.encode(original)
    emb_mixed = semantic_model.encode(mixed)
    similarity = torch.nn.functional.cosine_similarity(
        torch.tensor(emb_orig),
        torch.tensor(emb_mixed),
        dim=0
    )
    return similarity.item() > threshold

def augment_dataset(df: pd.DataFrame, text_column="bengali_text") -> pd.DataFrame:
    """Process DataFrame to generate code-mixed versions"""
    results = []

    for i in tqdm(range(0, len(df), BATCH_SIZE)):
        batch = df.iloc[i:i+BATCH_SIZE]
        batch_results = []

        for _, row in batch.iterrows():
            original_text = row[text_column]

            try:
                # Generate code-mixed version
                mixed_text = generate_code_mixed_sentence(original_text)

                # Validate semantics
                if validate_semantics(original_text, mixed_text):
                    batch_results.append({
                        "original": original_text,
                        "code_mixed": mixed_text,
                        "similarity_score": "valid"
                    })
                else:
                    batch_results.append({
                        "original": original_text,
                        "code_mixed": "VALIDATION_FAILED",
                        "similarity_score": "invalid"
                    })
            except Exception as e:
                batch_results.append({
                    "original": original_text,
                    "code_mixed": f"ERROR: {str(e)}",
                    "similarity_score": "error"
                })

        results.extend(batch_results)

    return pd.DataFrame(results)
```

```python
# ======================
# Fallback Mechanism
# (For failed validations)
# ======================

def contextual_replacement(bn_sentence: str, replace_ratio: float = 0.35) -> str:
    """Rule-based fallback for failed LLM generations"""
    tokens = bn_tokenizer.sentence_tokenize(bn_sentence)
    pos_tags = pos_tagger.bengali_pos(bn_sentence)

    # Identify replaceable content words
    content_words = []
    for token, pos in zip(tokens, pos_tags):
        if pos[1] in ["NN", "VB", "JJ"]:  # Nouns, Verbs, Adjectives
            content_words.append(token)

    # Select words to replace (based on ratio)
    n_replace = max(1, int(len(content_words) * replace_ratio))
    replace_candidates = content_words[:n_replace]

    # Simple dictionary (extend this with your domain terms)
    dictionary = {
        "বই": "book",
        "স্কুল": "school",
        "কলেজ": "college",
        "ছাত্র": "student",
        "শিক্ষক": "teacher",
        "পড়া": "study",
        "খেলা": "play",
        "দৌড়ানো": "run",
        "সুন্দর": "beautiful",
        "বড়": "big"
    }

    # Perform replacements
    mixed_tokens = []
    for token in tokens:
        if token in replace_candidates and token in dictionary:
            mixed_tokens.append(dictionary[token])
        else:
            mixed_tokens.append(token)

    return " ".join(mixed_tokens)

# Apply fallback to failed rows
failed_df = result_df[result_df['similarity_score'] != "valid"]
if not failed_df.empty:
    failed_df['code_mixed'] = failed_df['original'].apply(
        lambda x: contextual_replacement(x, REPLACE_RATIO)
    )
    full_df = pd.concat([valid_df, failed_df])
    full_df.to_csv("full_code_mixed_dataset.csv", index=False)
```

```python
# -*- coding: utf-8 -*-
import numpy as np
from bnlp import BengaliPOS, NLTKTokenizer
from nltk.corpus import stopwords
import nltk
import requests
import random
from sentence_transformers import SentenceTransformer

# Download required NLTK data
nltk.download('stopwords')

class BengaliEnglishCodeMixer:
    def __init__(self, replace_ratio=0.3, similarity_threshold=0.5):
        self.replace_ratio = replace_ratio
        self.similarity_threshold = similarity_threshold
        self.model = None
        self.english_embeddings = None
        self.english_words = None
        self.pos_tagger = None
        self.tokenizer = None
        self.bengali_stopwords = None

    def initialize(self):
        """Load models and resources"""
        print("Loading LaBSE embedding model...")
        self.model = SentenceTransformer('sentence-transformers/LaBSE')

        print("Initializing Bengali NLP tools...")
        self.pos_tagger = BengaliPOS()
        self.tokenizer = NLTKTokenizer()

        print("Loading English vocabulary...")
        self._load_english_vocab()

        print("Precomputing English embeddings...")
        self.english_embeddings = self.model.encode(self.english_words,
                                                  batch_size=128,
                                                  show_progress_bar=True)
        # Normalize embeddings for efficient cosine similarity
        self.english_embeddings = self.english_embeddings / np.linalg.norm(
            self.english_embeddings, axis=1, keepdims=True)

        print("Loading Bengali stopwords...")
        self._load_bengali_stopwords()

        print("System ready for code mixing")

    def _load_english_vocab(self, vocab_size=5000):
        """Load common English words"""
        url = "https://raw.githubusercontent.com/first20hours/google-10000-english/master/google-10000-english-usa.txt"
        response = requests.get(url)
        words = response.text.splitlines()
        self.english_words = words[:vocab_size]

    def _load_bengali_stopwords(self):
        """Load Bengali stopwords"""
        url = "https://raw.githubusercontent.com/stopwords-iso/stopwords-bn/master/stopwords-bn.txt"
        response = requests.get(url)
        self.bengali_stopwords = set(response.text.splitlines())

    def _should_replace(self, word, pos_tag):
        """Determine if a word should be considered for replacement"""
        # Define content POS tags: Noun, Verb, Adjective, Adverb
        content_pos = {"NN", "NNP", "NNC", "VM", "VAUX", "JJ", "RB"}

        return (pos_tag in content_pos and
                word not in self.bengali_stopwords and
                len(word) > 2 and  # Avoid short words
                not word.isdigit() and
                random.random() < self.replace_ratio)

    def _find_best_match(self, bengali_word):
        """Find closest English word using cosine similarity"""
        # Get Bengali word embedding
        word_embed = self.model.encode([bengali_word], show_progress_bar=False)[0]
        word_embed = word_embed / np.linalg.norm(word_embed)

        # Compute cosine similarity
        similarities = np.dot(self.english_embeddings, word_embed)
        best_idx = np.argmax(similarities)
        best_similarity = similarities[best_idx]

        return self.english_words[best_idx], best_similarity

    def code_mix_sentence(self, sentence):
        """Perform code mixing on a Bengali sentence"""
        if not self.model:
            self.initialize()

        # Tokenize and POS tag
        tokens = self.tokenizer.sentence_tokenize(sentence)
        pos_tags = self.pos_tagger.tag(sentence)

        # Handle token-POS alignment
        tokenized_sentence = [token for token, _ in tokens]
        pos_tags = [tag for _, tag in pos_tags] if len(pos_tags) == len(tokenized_sentence) else ["UNK"]*len(tokenized_sentence)

        mixed_tokens = []
        for i, (word, pos_tag) in enumerate(zip(tokenized_sentence, pos_tags)):
            if self._should_replace(word, pos_tag):
                try:
                    en_word, similarity = self._find_best_match(word)
                    if similarity >= self.similarity_threshold:
                        # Preserve Bengali suffixes for nouns/verbs
                        if pos_tag in {"NN", "NNP", "NNC", "VM"} and i < len(tokenized_sentence)-1:
                            next_pos = pos_tags[i+1]
                            if next_pos in {"PSP", "CCL", "SFX"}:
                                mixed_tokens.append(en_word + " ")  # Space for suffix attachment
                                continue
                        mixed_tokens.append(en_word)
                        continue
                except Exception as e:
                    print(f"Error processing '{word}': {str(e)}")

            # Keep original word if not replaced
            mixed_tokens.append(word)

        return " ".join(mixed_tokens)

    def code_mix_dataset(self, dataset):
        """Process a list of Bengali sentences"""
        return [self.code_mix_sentence(sentence) for sentence in dataset]

# Example Usage
if __name__ == "__main__":
    # Initialize code mixer (modify parameters as needed)
    mixer = BengaliEnglishCodeMixer(
        replace_ratio=0.4,       # 40% of eligible words replaced
        similarity_threshold=0.6  # Minimum similarity score
    )

    # Bengali input sentences
    bengali_sentences = [
        "আমি কাল বইমেলায় যাব",
        "এই গল্পটি খুব মজার ছিল",
        "কম্পিউটার বিজ্ঞান বিভাগে আমার ভর্তি হয়েছে",
        "আজ রাতের খাবারে মাছ থাকবে"
    ]

    # Perform code mixing
    mixed_sentences = mixer.code_mix_dataset(bengali_sentences)

    # Display results
    print("\nCode Mixed Output:")
    for orig, mixed in zip(bengali_sentences, mixed_sentences):
        print(f"Original: {orig}")
        print(f"Mixed   : {mixed}")
        print()
```

```python
# Check GPU availability
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

print("🔄 Loading models...")
embedding_model = SentenceTransformer('sentence-transformers/LaBSE', device=device)
tokenizer = Tokenizers()
pos_tagger = PosTagger()

# Load English vocabulary
with open("english_vocab.txt", "r") as f:
    english_vocab = [word.strip() for word in f.readlines() if 3 < len(word) < 15][:3000]  # Filtered

print("🧠 Precomputing English embeddings...")
english_embeddings = embedding_model.encode(english_vocab,
                                         batch_size=256,
                                         show_progress_bar=True,
                                         convert_to_tensor=True,
                                         device=device)
```

```python
def code_mix_sentence(sentence: str,
                     replace_ratio: float = 1,
                     pos_to_replace: list = ["NN", "VB", "JJ", "RB"]) -> str:
    """
    Args:
        sentence: Input Bengali sentence (e.g., "আমি বই পড়ি")
        replace_ratio: Probability of replacement (0-1)
        pos_to_replace: POS tags to replace ["NN", "VB", ...]
    Returns:
        Code-mixed sentence (e.g., "I বই read")
    """
    try:
        # Tokenize and POS tag
        words = tokenizer.bn_word_tokenizer(sentence)
        pos_tags = pos_tagger.tagger(sentence)

        if len(words) != len(pos_tags):
            return sentence  # Fallback

        mixed = []
        for word, (_, pos) in zip(words, pos_tags):
            if pos in pos_to_replace and random.random() < replace_ratio:
                # Get Bengali word embedding
                bn_embed = embedding_model.encode(word,
                                                 convert_to_tensor=True,
                                                 device=device)

                # Find most similar English word
                sims = cosine_similarity(
                    bn_embed.cpu().numpy().reshape(1, -1),
                    english_embeddings.cpu().numpy()
                )
                best_idx = np.argmax(sims)
                replacement = english_vocab[best_idx]

                # Preserve Bengali verb suffixes
                if pos == "VB" and word[-1] in ['া', 'ে', 'ো', 'ছ', 'ল']:
                    replacement += " " + word[-1]  # "eating" + " ছি"

                mixed.append(replacement)
            else:
                mixed.append(word)

        return " ".join(mixed)
    except Exception as e:
        print(f"Error processing: {sentence} - {str(e)}")
        return sentence

input_sentence = "আমার একটি সুন্দর বাগান আছে"

mixed = code_mix_sentence(input_sentence)
print(f"🔍 Original: {input_sentence}")
print(f"🌐 Mixed: {mixed}")
```

```python
import torch
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
from bnlp import BengaliPOS, BasicTokenizer, CleanText  # For accurate Bengali POS tagging
from functools import lru_cache
import re
from typing import List, Dict, Tuple
```

```python
# Configuration
MODEL_NAME = "facebook/nllb-200-distilled-1.3B"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
BATCH_SIZE = 1  # Optimal for most GPUs
SRC_LANG = "ben_Beng"
TGT_LANG = "eng_Latn"
ben_token = BasicTokenizer()
```

```python
# Initialize models
print("Loading models...")
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_NAME).to(DEVICE)
bn_pos = BengaliPOS()  # Bengali POS tagger

# POS tags to target (using Universal POS tags)
TARGET_POS_TAGS = {
    "NC",    # Common Noun (বই, গাছ)
    "NP",    # Proper Noun (ঢাকা, রহিম)
    "NV",    # Verbal Noun (পড়া, করা)
    "VM",    # Main Verb (যাই, খাই)
    "JJ",    # Adjective (সুন্দর, বড়)
    "JQ",    # Quantifier (কয়েক, অনেক)
    "RDF"    # Foreign Words (already likely to be English)
}  # bnlp uses Universal POS tags
```

```python
@lru_cache(maxsize=5000)  # Cache up to 5000 most frequent translations
def translate_word(word: str) -> str:
    """Cached single-word translation with error handling"""
    try:
        # Set source and target languages for the tokenizer
        tokenizer.src_lang = SRC_LANG
        tokenizer.tgt_lang = TGT_LANG

        inputs = tokenizer(word, return_tensors="pt").to(DEVICE)
        outputs = model.generate(
            **inputs,
            max_length=20,
        )
        return tokenizer.decode(outputs[0], skip_special_tokens=True)
    except Exception as e:
        print(f"Error translating '{word}': {str(e)}")
        return word
```

```python
def batch_translate(words: List[str]) -> List[str]:
    """Process words in batches for efficiency"""
    translations = []
    # Set source and target languages for the tokenizer
    tokenizer.src_lang = SRC_LANG
    tokenizer.tgt_lang = TGT_LANG

    for i in range(0, len(words), BATCH_SIZE):
        batch = words[i:i+BATCH_SIZE]
        inputs = tokenizer(batch, return_tensors="pt", padding=True).to(DEVICE)
        outputs = model.generate(
            **inputs,
            max_length=20,
        )
        translations.extend(tokenizer.batch_decode(outputs, skip_special_tokens=True))
    return translations
```

```python
def get_word_pos_pairs(sentence: str) -> List[Tuple[str, str]]:
    """Accurate Bengali POS tagging using bnlp"""
    # Tokenize preserving punctuation
    clean_text = CleanText(
        fix_unicode=True,
        unicode_norm=True,
        unicode_norm_form="NFKC",
        remove_url=True,
        remove_email=True,
        remove_emoji=True,
        remove_number=True,
        remove_digits=True,
        remove_punct=True
    )
    sentence = clean_text(sentence)
    tokens = ben_token(sentence)
    # Get POS tags for Bengali words (skip punctuation)
    words = [t for t in tokens]
    pos_tags = bn_pos.tag(" ".join(words))
    print(pos_tags)
    # Reconstruct with original tokens
    pos_map = {word: tag for word, tag in pos_tags}
    return [(token, pos_map.get(token, "PUNCT")) for token in tokens]
```

```python
def code_mix_sentence(sentence: str) -> str:
    """Optimized code-mixing with batch processing"""
    word_pos_pairs = get_word_pos_pairs(sentence)
    to_translate = []
    translate_indices = []

    # Identify words needing translation
    for i, (word, pos) in enumerate(word_pos_pairs):
        if pos in TARGET_POS_TAGS and re.match(r'^[\w]+$', word):
            to_translate.append(word)
            translate_indices.append(i)

    # Batch translate all target words
    if to_translate:
        translations = batch_translate(to_translate)
        # Update only single-word translations
        for idx, trans in zip(translate_indices, translations):
            if " " not in trans and trans != word_pos_pairs[idx][0]:
                word_pos_pairs[idx] = (trans, word_pos_pairs[idx][1])

    # Reconstruct sentence
    mixed = [wp[0] for wp in word_pos_pairs]
    return re.sub(r'\s+([?।!,])', r'\1', ' '.join(mixed))
```

```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from bnlp import BengaliPOS
import random


# --- Step 2: Initialize Models ---
embedding_model = SentenceTransformer("sentence-transformers/LaBSE")
bn_pos = BengaliPOS()

# --- Step 3: Define Replacement Rules via POS Tags ---
REPLACEABLE_TAGS = {
    'NC',  # Common Noun
    'NP',  # Proper Noun
    'NV',  # Verbal Noun
    'JJ',  # Adjective
    'VM',  # Main Verb (use cautiously)
    'JQ',  # Quantifier (e.g., "many" → "অনেক")
}

def get_replaceable_words(sentence_bn):
    """Extract words with replaceable POS tags."""
    tagged = bn_pos.tag(sentence_bn)
    return [
        (word, pos) for word, pos in tagged
        if pos in REPLACEABLE_TAGS
    ]

# --- Step 4: Context-Aware Code-Mixing ---
def code_mix_with_pos(sentence_bn, sentence_en, threshold=0.75, replace_prob=1):
    # Step 1: Get replaceable Bengali words and their POS
    replaceable_words = get_replaceable_words(sentence_bn)
    if not replaceable_words:
        return sentence_bn

    bn_words, bn_pos_tags = zip(*replaceable_words)

    # Step 2: Tokenize English translation (clean and lowercase)
    en_words = [
        word.lower() for word in sentence_en.split()
        if word.isalpha()
    ]
    if not en_words:
        return sentence_bn

    # Step 3: Compute embeddings
    bn_embeddings = embedding_model.encode(bn_words, convert_to_tensor=True)
    en_embeddings = embedding_model.encode(en_words, convert_to_tensor=True)

    # Step 4: Similarity matrix
    sim_matrix = cosine_similarity(bn_embeddings.cpu(), en_embeddings.cpu())

    # Step 5: Replace words with POS-based rules
    tokens = sentence_bn.split()
    replaced_tokens = []
    for token in tokens:
        if token in bn_words and random.random() < replace_prob:
            idx = bn_words.index(token)
            best_en_idx = np.argmax(sim_matrix[idx])
            best_similarity = sim_matrix[idx][best_en_idx]

            # POS-specific replacement logic
            pos = bn_pos_tags[idx]
            if (
                (best_similarity > threshold) and
                (pos in REPLACEABLE_TAGS) # and
                # Skip replacing verbs if English candidate isn't a verb (optional)
                # not (pos == 'VM' and not en_words[best_en_idx].endswith(('ing', 'ed')))
            ):
                replaced_tokens.append(en_words[best_en_idx])
            else:
                replaced_tokens.append(token)
        else:
            replaced_tokens.append(token)

    return " ".join(replaced_tokens)

# --- Step 5: Apply to DataFrame ---
df["code_mixed"] = df.apply(
    lambda row: code_mix_with_pos(row["tgt"], row["src"]),
    axis=1
)

# Save results
df.to_csv("code_mixed_pos_aware.csv", index=False)
```
