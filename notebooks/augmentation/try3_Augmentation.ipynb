```python
from bnlp import POS
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from google.cloud import translate_v2 as translate
import pandas as pd
import json
import os
from tqdm import tqdm
import time
```

```python
# %% [markdown]
"""
# Bengali-English Code Mixing Pipeline
**Features:**
1. POS-guided lexical replacement (Bengali → English)
2. Dynamic translation using Google Cloud API (for OOV words)
3. Fluency correction with BanglaT5
4. Caching for cost efficiency
"""

# %% [markdown]
"""
## Step 1: Install Dependencies
"""

# %%
# %% [markdown]
"""
## Step 2: Initialize Models
"""

# %%

# Initialize Bengali POS tagger
bn_pos = POS()

# Load BanglaT5 for fluency correction
tokenizer = AutoTokenizer.from_pretrained("csebuetnlp/banglat5_base")
model = AutoModelForSeq2SeqLM.from_pretrained("csebuetnlp/banglat5_base")

# %% [markdown]
"""
## Step 3: Google Cloud Setup (Run only once)
"""

# %%
# Upload your Google Cloud service account key (JSON)
from google.colab import files
uploaded = files.upload()
key_file = list(uploaded.keys())[0]
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = key_file

# Initialize Google Translate client
translate_client = translate.Client()

# %% [markdown]
"""
## Step 4: Core Functions
"""

# %%
# Load/save translation cache
def load_lexicon_cache(cache_path="lexicon_cache.json"):
    try:
        with open(cache_path, "r", encoding="utf-8") as f:
            return json.load(f)
    except FileNotFoundError:
        return {}

def save_lexicon_cache(cache, cache_path="lexicon_cache.json"):
    with open(cache_path, "w", encoding="utf-8") as f:
        json.dump(cache, f, ensure_ascii=False, indent=2)

LEXICON = load_lexicon_cache()

# Google Translate with rate limiting
def translate_word(bn_word, target_lang="en", max_retries=3):
    if bn_word in LEXICON:
        return LEXICON[bn_word]

    for _ in range(max_retries):
        try:
            result = translate_client.translate(bn_word, target_lang=target_lang)
            time.sleep(0.1)  # Rate limiting
            translated = result["translatedText"].lower()
            LEXICON[bn_word] = translated  # Cache result
            return translated
        except Exception as e:
            print(f"Retrying '{bn_word}': {e}")
            time.sleep(1)
    return None

# POS-guided code mixing
def pos_guided_code_mix(sentence, use_google=False):
    tagged = bn_pos.tag(sentence)
    mixed_words = []
    for word, pos in tagged:
        if pos in ["NN", "NNP", "JJ"]:  # Nouns/adjectives only
            if word in LEXICON:
                mixed_words.append(LEXICON[word])
            elif use_google:
                translated = translate_word(word)
                mixed_words.append(translated if translated else word)
            else:
                mixed_words.append(word)
        else:
            mixed_words.append(word)
    return " ".join(mixed_words)

# Fluency correction with BanglaT5
def refine_with_lm(text):
    input_ids = tokenizer.encode("Fix grammar: " + text, return_tensors="pt")
    outputs = model.generate(input_ids, max_length=50)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# %% [markdown]
"""
## Step 5: Process Your DataFrame
"""

# %%
# Assuming your DataFrame is named `df` with a column 'text'
def process_batch(df, text_column="text", use_google=True, save_path="output.csv"):
    tqdm.pandas(desc="Code-mixing progress")

    # Step 1: POS + Google Translate
    df["code_mixed"] = df[text_column].progress_apply(
        lambda x: pos_guided_code_mix(x, use_google=use_google)
    )

    # Step 2: Fluency correction (sample 20% to save time)
    sample_size = min(1000, len(df))  # Adjust based on your needs
    sample = df.sample(sample_size)
    sample["refined"] = sample["code_mixed"].progress_apply(refine_with_lm)

    # Save cache and results
    save_lexicon_cache(LEXICON)
    df.to_csv(save_path, index=False)
    return df

# Example usage:
# df_processed = process_batch(df, text_column="your_text_column")

# %% [markdown]
"""
## Step 6: Run Pipeline (Customize for Your DF)
"""

# %%
# Replace this with your actual DataFrame loading code
# Example dummy data
df = pd.DataFrame({
    "text": [
        "আমার বইটি টেবিলে আছে",
        "সে স্কুলে যায়",
        "এই অপ্রচলিত শব্দটি অনুবাদ করুন"
    ]
})

# Process all rows
df_processed = process_batch(df, text_column="text", use_google=True)

# Show results
print(df_processed.head())

# %% [markdown]
"""
## Optional: Pre-cache Frequent Words
"""

# %%
def pre_cache_words(texts, top_n=500):
    words = []
    for text in texts:
        tagged = bn_pos.tag(text)
        words.extend([word for word, pos in tagged if pos in ["NN", "NNP", "JJ"]])

    common_words = pd.Series(words).value_counts().head(top_n).index.tolist()
    for word in tqdm(common_words, desc="Pre-caching translations"):
        if word not in LEXICON:
            translate_word(word)
    save_lexicon_cache(LEXICON)

# Run on your dataset
# pre_cache_words(df["text"].tolist())
```

```python
# -*- coding: utf-8 -*-
"""Bengali-English Code Mixing with Local Translation Fallback.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1Yb4G3E5J9Y8Q7Q9Q9Q9Q9Q9Q9Q9Q9Q9
"""

# Step 1: Install all required packages
!pip install -q bnlp-toolkit transformers pandas tqdm requests torch sentencepiece

# Step 2: Import all necessary libraries
from bnlp import BengaliPOS
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
import pandas as pd
import json
import requests
from tqdm.auto import tqdm
import torch
import time
from collections import defaultdict
import warnings
warnings.filterwarnings('ignore')

# Initialize Bengali POS tagger
print("Loading Bengali POS tagger...")
bn_pos = BengaliPOS()

# Load local translation model (NLLB-200 Distilled 600M)
print("Loading translation models...")
device = 0 if torch.cuda.is_available() else -1
local_translator = pipeline(
    "translation",
    model="facebook/nllb-200-distilled-600M",
    tokenizer="facebook/nllb-200-distilled-600M",
    src_lang="ben_Beng",
    tgt_lang="eng_Latn",
    device=device
)

# Load BanglaT5 for fluency correction
banglat5_tokenizer = AutoTokenizer.from_pretrained("csebuetnlp/banglat5_small")
banglat5_model = AutoModelForSeq2SeqLM.from_pretrained("csebuetnlp/banglat5_small")

# Initialize translation cache
translation_cache = defaultdict(dict)
CACHE_FILE = "translation_cache.json"

def load_cache():
    try:
        with open(CACHE_FILE, 'r', encoding='utf-8') as f:
            return json.load(f)
    except (FileNotFoundError, json.JSONDecodeError):
        return defaultdict(dict)

def save_cache():
    with open(CACHE_FILE, 'w', encoding='utf-8') as f:
        json.dump(translation_cache, f, ensure_ascii=False, indent=2)

translation_cache = load_cache()

def translate_with_libretranslate(text, source="bn", target="en"):
    """Translate using LibreTranslate API"""
    try:
        response = requests.post(
            "https://libretranslate.de/translate",
            data={
                "q": text,
                "source": source,
                "target": target,
                "format": "text"
            },
            timeout=5
        )
        if response.status_code == 200:
            return response.json()["translatedText"]
    except Exception as e:
        print(f"LibreTranslate error: {str(e)}")
    return None

def translate_with_local(text, source_lang="ben_Beng", target_lang="eng_Latn"):
    """Translate using local NLLB model"""
    try:
        result = local_translator(text, max_length=100)
        return result[0]['translation_text']
    except Exception as e:
        print(f"Local translation error: {str(e)}")
    return None

def translate_word(word, source_lang="bn", target_lang="en"):
    """Smart translation with caching and fallback"""
    # Check cache first
    if word in translation_cache.get(source_lang, {}):
        return translation_cache[source_lang][word]

    # Try LibreTranslate first
    translated = translate_with_libretranslate(word, source_lang, target_lang)

    # Fallback to local model if API fails
    if not translated:
        translated = translate_with_local(word, f"{source_lang}_Beng", f"{target_lang}_Latn")

    # If we got a translation, clean and cache it
    if translated:
        translated = translated.lower().strip()
        translation_cache[source_lang][word] = translated
        save_cache()
        return translated

    return None

def pos_guided_code_mix(sentence, use_translator=True, replace_nouns=True, replace_adjectives=True):
    """Generate code-mixed sentence using POS tagging"""
    tagged = bn_pos.tag(sentence)
    mixed_words = []

    for token in tagged:
        word, pos = token["word"], token["pos"]

        # Determine if we should translate this word
        should_translate = (
            (replace_nouns and pos in ["NN", "NNP"]) or
            (replace_adjectives and pos == "JJ")
        )

        if should_translate and use_translator:
            translated = translate_word(word)
            mixed_words.append(translated if translated else word)
        else:
            mixed_words.append(word)

    return " ".join(mixed_words)

def refine_with_banglat5(text):
    """Improve fluency of code-mixed text"""
    input_text = f"Improve this Bengali-English code-mixed sentence: {text}"
    input_ids = banglat5_tokenizer.encode(input_text, return_tensors="pt")

    if device != -1:
        input_ids = input_ids.to(device)

    outputs = banglat5_model.generate(input_ids, max_length=100)
    return banglat5_tokenizer.decode(outputs[0], skip_special_tokens=True)

def process_batch(df, text_column="text", batch_size=100, use_translator=True):
    """Process a DataFrame in batches"""
    results = []

    for i in tqdm(range(0, len(df), batch_size), desc="Processing batches"):
        batch = df.iloc[i:i+batch_size]

        # Process each sentence in the batch
        batch_results = []
        for sentence in batch[text_column]:
            try:
                # Generate code-mixed version
                mixed = pos_guided_code_mix(sentence, use_translator=use_translator)

                # Refine a subset for quality checking
                if i % 5 == 0:  # Refine every 5th sentence
                    refined = refine_with_banglat5(mixed)
                    batch_results.append((sentence, mixed, refined))
                else:
                    batch_results.append((sentence, mixed, None))
            except Exception as e:
                print(f"Error processing sentence: {str(e)}")
                batch_results.append((sentence, sentence, None))

        results.extend(batch_results)
        time.sleep(1)  # Rate limiting

    # Create result DataFrame
    result_df = pd.DataFrame(results, columns=["original", "code_mixed", "refined"])
    return result_df

# Example usage
if __name__ == "__main__":
    # Sample data
    sample_data = [
        "আমার বইটি টেবিলে আছে",
        "সে স্কুলে যায়",
        "এই অপ্রচলিত শব্দটি অনুবাদ করুন",
        "আমার কম্পিউটারটি নতুন",
        "তিনি ডাক্তার হিসাবে কাজ করেন"
    ]

    df = pd.DataFrame({"text": sample_data})

    print("\nStarting code mixing process...")
    result_df = process_batch(df, text_column="text", batch_size=2)

    print("\nResults:")
    print(result_df)

    # Save final cache
    save_cache()
    print("\nTranslation cache saved.")

    # Save results to CSV
    result_df.to_csv("code_mixed_results.csv", index=False)
    print("Results saved to 'code_mixed_results.csv'")
```
